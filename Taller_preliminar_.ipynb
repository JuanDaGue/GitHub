##Superposicion de ondas

Superposición de ondas. Una onda clásica en $1D$ con número de onda fijo $k_0$ y frecuencia
fija $\omega_0$ es de la forma
$$
	\Psi(x, t) = e^{i(k_0x−\omega_0 t)}
$$
que implica tener una longitud de onda fija $\lambda = 2\pi/k_0$ y una frecuencia temporal fija $\nu =\omega /2\pi$ dando lugar a ondas planas de rango infinito.
Sin embargo, por superposición de ondas podemos generar confinamientos espaciales o temporales.
Si lugar de tomar el vector de onda central $k_0$, se toman dos próximos $[k_0−\Delta k, k_0+\Delta k]$
asociados respectivamente con las frecuencias $[\omega_0 − \Delta\omega, \omega _0 + \Delta\omega]$, se puede construir la superposición
de dos ondas y de su interferencia. Indague analíticamente cuál es la forma
compacta de esta nueva onda de superposición así como su velocidad de fase y velocidad de
grupo. Dibuje cualitativamente la nueva onda de superposición.

```

############# Función de Onda ##########################

def Psi(x,t,k,w):
  return np.e**(1j*(k*x-w*t))
```

```
#Condiciones iniciales
k0=1
w0=2

#Arrays
##Diferentials
Dk=np.arange(1,10,1)
Dw=np.arange(1,10,1)

#Time and possition
t=np.linspace(0,50,500)
x=np.linspace(0,50,500)
# Wave function 
Y=Psi(x,t,k0,w0)

```

```
#####################################################################
#Graphics of the superpotition wave
##################################################################
#Cosntantes
A=0.5
B=0.5

#Funcion superpuesta 

```

```
from ipywidgets import *

def Sp(n,m):
  Yt=A*Psi(x,t,k0+Dk[n],w0+Dw[m])+B*Psi(x,t,k0-Dk[n],w0-Dw[m])
  plt.plot(t,Yt.real,'k')
  plt.plot(t,Y.real,"--",color='darkred')
  plt.plot(t,-Y.real,"--",color='darkred')
  plt.grid()
  plt.xlim(0,8*np.pi)

interact(Sp,n=(0,len(Dk)-1),m=(0,len(Dw)-1))
```

#Particula en una Caja  $3D$

para esto pirmero construimos la particula en una dimension


```
L2=10.0
x2=np.linspace(0,L2,100)
def y2(x,n,L):
  return np.sqrt(2/L)*np.sin((n*np.pi*x)/L)
def yn(n,L2):
  plt.figure(figsize=(7,7))
  plt.plot(x2,y2(x2,n,L2),"darkblue",label=r'Función de onda $\Psi$ para n='+str(n))
  plt.title("Partícula en una caja 1D")
  #plt.xlim(0,L/5)
  plt.grid()
  plt.legend(loc='lower right')
interact(yn,n=(1,10),L2=(1,20))
```

```
############# particula en el espacio R^3
def Y_n(r,n,L):
  return np.sqrt(8/L[0]*L[1]*L[2])*np.sin((n[1]*np.pi*r[1])/L[1])*np.sin((n[0]*np.pi*r[1])/L[0])*np.sin((n[2]*np.pi*r[2])/L[2])
```

```
############Vector de posiction#######################
r=np.array([x2,x2,x2])
n3=np.array([2,2,2])
L3=np.array([L2,1*L2,1*L2])
```

```
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

```

```
def D_n(nx,ny,nz):
  fig = plt.figure(figsize=(10,10))
  ax = fig.add_subplot(111, projection='3d')
  n3=np.array([nx,ny,nz])
  # Make data.
  X, Y  = np.meshgrid(r[0], r[1])
  R=np.array([X,Y,Y])

  Z=Y_n(R,n3,L3)**2

  # Plot the surface.
  surf = ax.plot_surface(X, Y, Z, cmap=cm.inferno_r,
                        linewidth=0,)

  # Customize the z axis.

  #ax.set_zlim(-10, 10)
  #ax.set_zlim(0, (L3[2])**2)
  #ax.zaxis.set_major_locator(LinearLocator(10))
  #ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
  ax.set_title(r"Partícula en una caja $3 D$",fontsize=(20))
  ax.set_xlabel("X label")
  ax.set_ylabel("Y label")
  ax.set_zlabel("Z label")

  # Add a color bar which maps values to colors.

  fig.colorbar(surf, shrink=0.5, aspect=30)
interact(D_n,nx=(1,10),ny=(1,10),nz=(1,10))
```
